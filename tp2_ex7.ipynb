### Resumo de Notícia Utilizando Exemplos em Prompts

Usando a API Gemini, crie um notebook que utilize prompts para resumir uma notícia (escolha da página principal de um portal de notícias e copie no notebook). O prompt deve solicitar um resumo dessa notícia usando o princípio de Exemplos para guiar a resposta do LLM. Teste o código e avalie a qualidade do resumo gerado.

```python
import google.generativeai as genai
from google.colab import userdata
```

```python
# Notícia copiada do portal de notícias (exemplo de notícia)
link_noticia = 'https://www.autodata.com.br/noticias/2024/11/07/volkswagen-t-cross-e-virtus-ganham-versao-sense-de-entrada/80395/'

noticia = """
São Paulo – A Volkswagen ampliou a sua lista de modelos com a configuração Sense, de entrada, que chegou ao sedã Virtus
e ao SUV T-Cross, e já era ofertada nos modelos Polo e Nivus.
O novo T-Cross conta com uma versão Sense atualizada e oferece kit multimídia VW Play com tela sensível ao
toque de 10,1 polegadas, faróis e lanternas em led, quadro de instrumentos digital de oito polegadas, ar-condicionado,
motor 1.0 turbo de 130 cv de potência e câmbio automático de seis marchas.
Já o Virtus Sense é equipado com motor turbo 170 TSI que entrega 116 cv de potência, acoplado ao câmbio manual
de seis marchas, sistema multimídia Media Plus II com tela sensível ao toque, volante multifuncional, quadro de
instrumentos digital de oito polegadas, ar-condicionado, faróis e lanternas em led, seis airbags, assistente de
partida em rampa, dentre outros.

Veja abaixo o preço das novas versões:

Virtus Sense – R$ 104 mil
T-Cross Sense – R$ 120 mil
"""
```

```python
# Criando o prompt utilizando o princípio de Exemplos (few-shot prompting)
prompt = f"""
resuma a <NOTICIA> em até 2 sentenças com os principais pontos da noticia.
aqui esta um exemplo:

<NOTICIA>
"A concorrencia no mercado de automoveis vem crescendo internamento no mercado brasileiro, trazendo novas marcas para
disputar o marketshare no mercado como a chinesa BYD. Modelos mais vendidos hoje é o POLO e em seguida HB20.

<RESUMO>
"Montadora chinesa BYD tem pela frente os modelos POLO e HB20 como concorrentes.

Agora, faça o resumo da noticia abaixo:
<NOTICIA>
{noticia}
###
<RESUMO>
"""
```

```python
# Definindo a API_KEY
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key = GOOGLE_API_KEY)
```

```python
# Definindo o modelo e resumindo a noticia
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content(prompt)
```

```python
print(len(noticia),len(response.text))
```

```python
# Exibindo a resposta gerada
print("Resumo gerado pelo LLM:")
print(response.text.replace("#","").replace(". ",".\n"))
```
