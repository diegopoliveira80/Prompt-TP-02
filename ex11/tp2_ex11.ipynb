```python
import yaml

with open("/content/drive/MyDrive/Colab Notebooks/Prompt/TP 02/ex11/roteiro.yaml") as stream:
    try:
        config = yaml.safe_load(stream)
    except yaml.YAMLError as exc:
        print(exc)

config
```

```python
import google.generativeai as genai
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key = GOOGLE_API_KEY)
```

```python
for roteiros in config['roteiro'].keys():
  model = genai.GenerativeModel('gemini-1.5-flash')
  response = model.generate_content(config['roteiro'][roteiros])


  # Exibindo a resposta gerada
  print(f"Resumo gerado pelo LLM roteiro {roteiros.upper()}:")
  print(response.text.replace("**","").replace(". ","\n"))
```
