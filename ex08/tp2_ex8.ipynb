### Identificação de Entidades em Notícias com LLM

Uma consultoria lhe contratou para automatizar a descoberta de pessoas mencionadas em notícias de jornal. Com base no Exercício 7, escolha 3 notícias e monte uma aplicação com um prompt para o LLM identificar menções a diferentes entidades em cada notícia (como pessoas, órgãos públicos, empresas…). Implemente um notebook para testar o código usando a API do Gemini. Descreva o prompt, seus resultados e avalie se a resposta do modelo atende às expectativas da consultoria.

```python
import google.generativeai as genai
from google.colab import userdata
import json
```

```python
# Definindo a API_KEY
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')
genai.configure(api_key = GOOGLE_API_KEY)
```

```python
noticias= "/content/drive/MyDrive/Colab Notebooks/Prompt/TP 02/ex08/noticias.json"
with open(noticias, "r") as file:
  dados_noticias = json.load(file)
```

```python
def prompt(dados):
  # Notícia copiada do portal de notícias (exemplo de notícia)
  for i in range(len(dados)):

    # Criando o prompt utilizando o princípio de Exemplos (few-shot prompting)
    prompt = f"""
    Encontre na <NOTICIA> pessoas mencionadas menções a diferentes entidades em cada notícia
    (como pessoas, órgãos públicos, empresas…).
    aqui esta um exemplo:

    <NOTICIA>
    "O governo federal anunciou hoje que o Ministério da Saúde, liderado pela Ministra Ana Costa, está firmando uma
    parceria com a empresa de tecnologia Infotech Solutions para desenvolver um sistema de monitoramento de saúde pública.
    A nova plataforma visa melhorar a coleta de dados em tempo real e permitir uma resposta mais rápida a surtos de doenças
    no país. Além disso, o presidente da Infotech, João Pereira, declarou que a colaboração com o governo federal é um passo
    importante para a empresa expandir suas operações no setor público. O Ministério da Saúde, por meio da sua Secretaria
    de Vigilância em Saúde, também planeja realizar uma série de workshops com representantes da Organização Mundial da
    Saúde (OMS) para discutir o uso de novas tecnologias no combate a epidemias.

    <RESULTADO>
    "Pessoas mencionadas:

        Ministra Ana Costa – Líder do Ministério da Saúde.
        João Pereira – Presidente da empresa Infotech Solutions.

    Entidades mencionadas:

        Ministério da Saúde – Órgão público responsável pela saúde no Brasil.
        Infotech Solutions – Empresa de tecnologia que estará em parceria com o governo.
        Organização Mundial da Saúde (OMS) – Instituição internacional que atuará em conjunto com o Ministério da Saúde.


    Agora, faça a pesquisa:
    <NOTICIA>
    {dados[i][f'noticia_{i+1}']}
    ###
    <RESULTADO>
    """


    # Definindo o modelo e resumindo a noticia
    model = genai.GenerativeModel("gemini-1.5-flash")
    response = model.generate_content(prompt)

    # Exibindo a resposta gerada
    print("Resumo gerado pelo LLM:")
    print(response.text.replace("#","").replace(". ",".\n").replace("**",""))

prompt(dados_noticias)

```
